### 1. Ler e exibir um arquivo CSV usando pandas

```
path = '/FileStore/tables/retail.csv'
df = spark.read.csv(path, header=True,sep=',')
df = df.toPandas()

df
```

### 2. Selecionar colunas específicas de um DataFrame usando pandas

```
# vamos supor que quero selecionar as colunas stock_code e unit_price
df[['stock_code','unit_price']]
```

### 3. Filtrar linhas com base em uma condição usando pandas

```
# Método 1: filtro
df.filter(like='id')
```

```
# Método 2:
df[df['customer_id'] == '12680'] # filtar dois itens df[(df['customer_id'] == '12680') & (df['unit_price'] == '0,29')]
```

```
# Método 3: query
df.query('customer_id =="12680"') # para filtar mais de um item df.query('customer_id == "12680" and unit_price == "0,29"')
```

### 4. Agrupar dados com base em uma coluna usando pandas

```
df.groupby('customer_id').agg(n=('stock_code', 'count')).reset_index()
```

### 5. Usar a função max em alguma coluna do DataFrame

```
df['quantity'].max()
```
