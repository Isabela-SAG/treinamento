```
soma = 0 # valor resultante da soma
for i in range (1, 101): # tem que ser até 101 para ele contar até 100
    soma += i # 
print("A soma de todos os números de 1 até 100 é", soma)
```

```
soma_pares = 0 # soma de todos os pares
for i in range(1, 101):
    if i % 2 == 0: # condição para ser par
        soma_pares += + i
print("A soma de todos os números pares de 1 até 100 é", soma_pares)
```

```
soma_impar = 0 # soma de todos os pares
for i in range(1, 101):
    if i % 2 != 0: # condição para ser impar
        soma_impar += + i
print("A soma de todos os números ímpares de 1 até 100 é", soma_impar)
```

```
soma == soma_pares + soma_impar # tem que ser True
```

```
lista = [-2, 2, 'celular', 10, -99, False]
lista
```

```
'''
analisando a lista, vemos que todas as variáveis numéricas são inteiras, contudo, devemos considerar que a lista
é mutável e podem inserir variáveis do tipo float (e não serem inteiras) e, por isso, vamos generalizar o máximo 
possível
'''

n = len(lista) # tamanho da lista
soma = 0 # soma deste ex
for i in range(n):
    if type(lista[i]) == int: # primeiro devemos verificar o tipo de dado
        if lista[i] > 0: # depois verificamos se é positivo
            soma += lista[i]
print("A soma é:", soma)
```
