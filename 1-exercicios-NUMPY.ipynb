### 1. Criar um array com 10 números inteiros aleatórios entre 1 e 50 e calcular a média e o desvio padrão

```
# https://numpy.org/doc/stable/reference/random/generated/numpy.random.randint.html
np_array = np.random.randint(1,51,10)
np_array
```

```
# média
print("Média:" ,np.mean(np_array))

# desvio padrão
print("Desvio padrão: ", np.std(np_array))
```

### 2. Criar um array de 10 números aleatórios e calcular a raiz quadrada de todos os valores menores ou iguais a 0,5

```
arr = np.random.rand(10)

print("Array original:\n",arr)

arr[arr<=0.5] = np.sqrt(arr[arr<=0.5])
print("\nRaiz quadrados dos valores <= 0.5\n", arr)
```

### 3. Criar uma matriz 3D de tamanho 3x3x3 com valores aleatórios e exibir o valor máximo de cada linha de cada matriz 2D dentro dessa matriz 3D

```
arr = np.random.rand(3, 3, 3)
arr
```

```
for i in range(len(arr)):
  print("\n%dª Matriz 2D" %(i+1))
  for j in range(len(arr[i])):
    maximo = arr[i][j].max()
    print("%dª Linha da Matriz %d - Valor máximo: %f " %(j+1,i+1,maximo))
    
```

### 5. Criar uma matriz de tamanho 4x4 com valores aleatórios e calcular a média e o desvio padrão das diagonais da matriz

```
matriz = np.random.rand(4,4)
matriz
```

```
matriz[0,-1], matriz[1,-2], matriz[2,-3], matriz[3,-4]
```

```
diag_princ = np.diag(matriz)
diag_sec = np.diag(np.fliplr(matriz))
media = np.mean([diag_princ,diag_sec])
std_matriz = np.std([diag_princ,diag_sec])

print("Diagonal principal:",diag_princ)
print("Diagonal secundária",diag_sec)
print("Média das diagonais", media)
print("Desvio padrão das diagonais:", std_matriz)
```
